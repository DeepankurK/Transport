```python
import pandas as pd
import numpy as np
```

```python
r1=pd.read_csv("../data/temp.csv",header=None)
r2=pd.read_csv("../data/temp2.csv",header=None)
for i in range(0,r1.shape[0]):r1.iloc[i,1]=r1.iloc[i,1].split('[')[-1].split(']')[0] if '[' in r1.iloc[i,1] else r1.iloc[i,1]
r1=r1.apply(pd.to_numeric)
r2=r2.apply(pd.to_numeric)
```

```python
#graph -1 selection
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')


x=np.array([0,1,2,3,4,5])
y=np.array([0,1,2,3,4,5])
a,b=np.meshgrid(x,y)
x,y=a.ravel(),b.ravel()

z=[0]*len(x)

for i in range(0,6):
    t=r1[i::6].iloc[:,2].value_counts()
    
    for j in t.index:
        z[i+6*j]=t.loc[j]

#print(z)
bottom = np.zeros_like(z)
dx = dy= 1

for i in range(0,6):
    print(sum(z[i::6]))


ax1.bar3d(y, x, bottom, dx,dy, z, shade=True)

ax1.set_ylabel('Number of Iterations of the Algorithm')
ax1.set_xlabel('Journey Class of the Flight')
ax1.set_zlabel('No. of flights')
ax1.set_yticklabels(['0',' Org','  Pred','  15','  20','  30','  60'],rotation=-45,va='baseline',ha='center')
ax1.set_xticklabels(['-1','0','1','2','3','4','5'],rotation=-15,va='center',ha='left')
#ax1.set_yticklabels([i for i in range(-1,6)],rotation=+45,va='center',ha='right')
ax1.set_title('Journey Class of the Flight vs Number of Iterations of the Algorithm')

plt.show()
a=np.reshape(z,(6,6))
print(a)
```

```python
#graph 1-selection
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')


x=np.array([0,1,2,3,4])
y=np.array([0,1,2,3,4,5])
a,b=np.meshgrid(x,y)
x,y=a.ravel(),b.ravel()
#print(x,y)
z=[0]*len(x)

for i in range(0,5):
    t=r2[i::5].iloc[:,2].value_counts()
    #print(t)
    for j in t.index:
        z[i+5*j]=t.loc[j]

#print(z)
bottom = np.zeros_like(z)
dx = dy= 1

for i in range(0,5):
    print(sum(z[i::5]))


ax1.bar3d(y, x, bottom, dx,dy, z, shade=True)

#ax1.set_ylabel('Different Error Metrics')
ax1.set_xlabel('Journey Class of the Flight')
ax1.set_zlabel('No. of flights')
ax1.set_yticklabels(['0','Org\n','  Metric:1','  Metric:2','  Metric:3','Metric:4'],rotation=-30,va='center',ha='left')
ax1.set_xticklabels(['-1','0','1','2','3','4','5'],rotation=-15,va='top',ha='left')
#ax1.set_yticklabels([i for i in range(-1,6)],rotation=+45,va='center',ha='right')
ax1.set_title('Journey Class of the Flight vs Different Error Metrics')

plt.show()
#print(z)
a=np.reshape(z,(6,5))
print(a)
```

```python
#graph 2
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')


x=np.array([0,1])
y=np.array([0,1,2,3,4,5])
a,b=np.meshgrid(x,y)
x,y=a.ravel(),b.ravel()
print(x,y)
z=np.zeros_like(x)
bottom = np.zeros_like(z)

t=r1.copy(deep=True)
#print(t)

t[1::6].iloc[:,1]=t[1::6].iloc[:,1].values-t[0::6].iloc[:,1].values
t[2::6].iloc[:,1]=t[2::6].iloc[:,1].values-t[0::6].iloc[:,1].values
for j in range(0,6):
    r=t[1::6]
    z[2*j]=abs(r[r.iloc[:,2]==j].iloc[:,1].median())
    
for j in range(0,6):
    r=t[2::6]
    z[1+2*j]=0 if len(r[r.iloc[:,2]==j].iloc[:,1])==0 else abs(r[r.iloc[:,2]==j].iloc[:,1].median())
    
print(z)

bottom = np.zeros_like(z)
dx = dy= 1


ax1.bar3d(x,y, bottom, dx,dy, z, shade=True)

ax1.set_xlabel('Effect of the Algorithm')
ax1.set_ylabel('Journey Class of the Flight')
ax1.set_zlabel('MAE of Taxi-Out Time')
ax1.set_xticklabels(['0','','Pred','','Optimised'],rotation=-15,va='baseline',ha='center')
ax1.set_yticklabels(['-1','','0','1','2','3','4','5'],rotation=-15,va='center',ha='left')
#ax1.set_yticklabels([i for i in range(-1,6)],rotation=+45,va='center',ha='right')
ax1.set_title('MAE of Taxi-Out Time vs Number of Iterations of the Algorithm vs Journey Class of the Flight')
plt.show()
a=np.reshape(z,(6,2))
print(a)
```

```python
#graph 2
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')


x=np.array([0,1,2,3])
y=np.array([0,1,2,3,4,5])
a,b=np.meshgrid(x,y)
x,y=a.ravel(),b.ravel()
print(x,y)
z=np.zeros_like(x)
bottom = np.zeros_like(z)

t=r2.copy(deep=True)
#print(t)

t[1::5].iloc[:,1]=t[1::5].iloc[:,1].values-t[0::5].iloc[:,1].values
t[2::5].iloc[:,1]=t[2::5].iloc[:,1].values-t[0::5].iloc[:,1].values
t[3::5].iloc[:,1]=t[3::5].iloc[:,1].values-t[0::5].iloc[:,1].values
t[4::5].iloc[:,1]=t[4::5].iloc[:,1].values-t[0::5].iloc[:,1].values
for j in range(0,6):
    r=t[1::5]
    z[4*j]=0 if len(r[r.iloc[:,2]==j].iloc[:,1])==0 else abs(r[r.iloc[:,2]==j].iloc[:,1].median())
    
for j in range(0,6):
    r=t[2::5]
    z[1+4*j]=0 if len(r[r.iloc[:,2]==j].iloc[:,1])==0 else abs(r[r.iloc[:,2]==j].iloc[:,1].median())
    
for j in range(0,6):
    r=t[3::5]
    z[2+4*j]=0 if len(r[r.iloc[:,2]==j].iloc[:,1])==0 else abs(r[r.iloc[:,2]==j].iloc[:,1].median())
    
for j in range(0,6):
    r=t[2::5]
    z[3+4*j]=0 if len(r[r.iloc[:,2]==j].iloc[:,1])==0 else abs(r[r.iloc[:,2]==j].iloc[:,1].median())

print(z)

bottom = np.zeros_like(z)
dx = dy= 1


ax1.bar3d(y,x, bottom, dx,dy, z, shade=True)

ax1.set_ylabel('Effect of the Algorithm')
ax1.set_xlabel('Journey Class of the Flight')
ax1.set_zlabel('MAE of Taxi-Out Time')
ax1.set_yticklabels(['0','','Metric:1','','Metric:2','','Metric:3','','Metric:4'],rotation=-15,va='baseline',ha='center')
ax1.set_xticklabels(['-1','','0','1','2','3','4','5'],rotation=-15,va='center',ha='left')
#ax1.set_yticklabels([i for i in range(-1,6)],rotation=+45,va='center',ha='right')
ax1.set_title('MAE of Taxi-Out Time vs Number of Iterations of the Algorithm vs Journey Class of the Flight')
plt.show()
a=np.reshape(z,(6,4))
print(a)
```
